# GENERATE RANDOM NUMBER and Play JACKPOT GAME

```python
import random # This Lirary Generates Random Numbers within a Range
```

```python
jackpot = random.randint(1,100) # Generate Random Number within the Range of 1 to 100 

guess = int(input("Guess a Number and Enter the Value : "))
counter = 1

while guess !=jackpot :
  
        
        
    if guess < jackpot:
        print("Guess Higher")
        guess = int(input("Guess a Number and Enter the Value : "))  
        counter += 1
    elif guess > jackpot :
        print("Guess Lower")
        guess = int(input("Guess a Number and Enter the Value : "))  
        counter += 1
print("\nCorrect Guess")

if counter <= 3 :
    print("You Won Jackpot")
else:
    print("\nHard Luck, Try Another Day")       
print("\nYou Took" , counter , " Guesses")

```

## BREAK , CONTINUE , PASS statements

### BREAK Statement

* #### BREAK Statement Immediately Terminates the Current Loop where the Break Statement Condition is Met
* #### For Example ,
  ####   if i == 20 :
  ####     break;
* #### Once i = 20 , the Loop will Terminate and it will get out of the Loop

```python
for i in range( 1 , 10 ):
    if i == 5 :
        break;
    print(i)
```

#### LINEAR SEARCHING EXAMPLE ( Using BREAK )

```python
# LINEAR SEARCHING in a Big List
# Searching for 111 in the list
numbers = [
    3, 8, 12, 19, 24, 31, 37, 42, 48, 53,
    57, 61, 66, 72, 75, 81, 86, 90, 94, 99,
    102, 105, 108, 110, 111, 118, 124, 131, 139, 147,
    156, 163, 171, 184, 192, 205, 217, 229, 241, 256,
    268, 279, 291, 314, 336, 359, 387, 421, 456, 498
]
counter = 1 

for i in numbers :
    if i == 111 :
        break;
    counter += 1 
print( 111 , " is at the Position " , counter + 1 )
```
